```python
import folium
```

```python
m = folium.Map(location=[7.8731, 80.7718])

m
```

```python
folium.Circle(
    radius=500000,
    location=[7.8731, 80.7718],
    popup='Sri Lanka',
    color='#3186cc',
    fill=True,
    fill_color='blue'
).add_to(m)

folium.Circle(
    radius=500000,
    location=[1.3733, 32.2903],
    popup='Uganda',
    color='#3186cc',
    fill=True,
    fill_color='brown'
).add_to(m)

folium.Circle(
    radius=500000,
    location=[52.3555, 1.1743],
    popup='England',
    color='#3186cc',
    fill=True,
    fill_color='red'
).add_to(m)

folium.Circle(
    radius=500000,
    location=[71.7069, 42.6043],
    popup='Greenland',
    color='#3186cc',
    fill=True,
    fill_color='yellow'
).add_to(m)


m
```
