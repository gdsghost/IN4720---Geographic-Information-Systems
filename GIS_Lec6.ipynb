```python
import geopandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import osmnx as ox
```

```python
uwa_p = ox.geocode_to_gdf("Uva province, Sri Lanka")
uwa_p.plot();
plt.title("Uva Province");
```

```python
ax = uwa_p.plot(edgecolor = "0.1");
badulla_dist = ox.geocode_to_gdf("Badulla district, Sri Lanka")
badulla_dist.plot(ax = ax, edgecolor = "0.1", color = "yellow")
plt.title("Badulla District")
```

```python
badulla_drive_nx = ox.graph_from_place("Badulla district, Sri Lanka", network_type= "drive")
badulla_drive_nx
```

```python
ox.plot_graph(badulla_drive_nx)
```
