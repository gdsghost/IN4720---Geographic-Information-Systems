```python
import geopandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import files
data_load = files.upload()
```

```python
df = pd.read_csv('Data_SL.csv', index_col = 0)
df
```

```python
df['geometry'] = geopandas.GeoSeries.from_wkt(df['geometry'])
geo_df = geopandas.GeoDataFrame(df, crs = "EPSG:4326")
geo_df
```

```python
geo_df.plot(column = 'population', legend = 'true');
```

```python
temp_today = [30,29,26,29,25,29,29,31,30]
geo_df['temp_today'] = temp_today
geo_df
```

```python
rain_today = [29, 35, 45, 23, 70, 29, 34, 14, 5]
geo_df['rain_today'] = rain_today
geo_df
```

```python
geo_df['p_density'] = geo_df['population'] / geo_df['area']
geo_df
```

```python
geo_df.plot(column = 'p_density', legend = 'true');
```

```python
geo_df.plot(column = 'p_density', legend = 'true', legend_kwds={'label': "Population Density by District", 'orientation': "vertical"});
```
